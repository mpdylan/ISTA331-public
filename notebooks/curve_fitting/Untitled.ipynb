```python
bg = pd.read_csv('bluegills.csv', sep = '\t')
bg.head()
```

```python
bg.plot('age', 'length', kind = 'scatter')
```

```python
x_array = sm.add_constant(bg['age'])
model = sm.OLS(bg['length'], x_array)
result = model.fit()
result.summary()
```

```python
x_array = np.column_stack([bg['age'] ** i for i in range(3)])
model = sm.OLS(bg['length'], x_array)
result = model.fit()
result.summary()
```
